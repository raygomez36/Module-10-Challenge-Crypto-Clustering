```python
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#Creating a Dataframe from Crypto Market Data Sheet
df = pd.read_csv('crypto_market_data.csv', index_col='coin_id')
crypto_df = df.iloc[:, 2:9].values
crypto_df
```

```python
#Generate Summary Statistics
df.describe()
```

```python
df.head()
```

```python
#Using the Elbow Method
crypt = []
k = list(range(1, 11))

#Running the KMeans Algorithim
for i in k:
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(crypto_df)
    crypt.append(kmeans.inertia_)

    
elbow_data = {'k': k, 'crypt': crypt}
k_df_elbow = pd.DataFrame(elbow_data)

k_df_elbow.head()
#Plotting Results
#k_elbow = (plt.plot(range(1, 11), crypt),
#plt.title('Elbow Method'),
#plt.xlabel('Number of Clusters'),
#plt.ylabel('Crypt'),
#plt.show)
#k_elbow
```

```python
k_elbow = k_df_elbow.hvplot.line(

    x='k',
    y='crypt',
    title='K-Means Elbow Curve',
    xticks='k'
)
k_elbow
```

```python
# Define the model with 3 clusters
model = KMeans(n_clusters = 3, random_state = 0)

# Fitting the model
model.fit(df)

# Making the predictions
cr = model.predict(df)

# Creating a copy of the DataFrame
crypto_predictions_df = df.copy()

#Adding a class column with labels
crypto_predictions_df['Crypto Predict'] = cr

```

```python
k_crypto = crypto_predictions_df.hvplot.scatter(

    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Crypto Predict'
)
k_crypto
```
